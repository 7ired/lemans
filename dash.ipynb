```python
import pandas as pd
import re
```

```python
df = pd.read_csv('data.csv', delimiter=';')
```

```python
df = df.dropna(how='any')
```

```python
df.head()
```

```python
df['Drivers'][0]
```

```python
df['Drivers'][0].split('>')
```

```python
df['Drivers'] = df['Drivers'].apply(lambda x : x.split('>')[1])
```

```python
df['Drivers']
```

```python
df['Drivers'][3]
```

```python
df['Drivers'] = df['Drivers'].apply(lambda x : x.rstrip('[]a]'))
```

```python
df['Drivers']
```

```python
regex = r'\(([^)]+)\)'
```

```python
driver_country_codes = df['Drivers'].str.findall(regex)
team_country_codes = df['Team'].str.findall(regex)
```

```python
driver_country_codes.head()
```

```python
team_country_codes.head()
```

```python
df['Driver_nationality'] = driver_country_codes
df['Team_nationality'] = team_country_codes
```

```python
df['Drivers'] = df['Drivers'].apply(lambda x: x.split('(')[0])
```

```python
df['Team'] = df['Team'].apply(lambda x : x.split('(')[0])
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['Mi'].unique()
```

```python
df['Mi'] = df['Mi'].str.replace(',', '').astype(float)
```

```python
df.drop(columns=['Ref'], inplace=True)
```

```python
df.tail()
```

```python
df['Series'].unique()
```

```python
df['Series'] = df['Series'].apply(lambda x : x.split('T')[0])
```

```python
df.tail()
```

```python
df.rename(columns={'Car and engine make' : 'Car'}, inplace=True)
```

## feature engineering

```python
df.tail()
```

```python
df['Average_speed_kmh'] = round(df['Km']/24, 2)
```

```python
df['Average_speed_mph'] = round(df['Mi']/24, 2)
```

```python
df.tail()
```

```python
df['Car'] = df['Car'].apply(lambda x : x.rstrip('[]g]'))
```

```python
df['Average_lap_time'] = round(24*3600/60/df['Laps'], 2)
```

```python
df.tail()
```

```python
df['Tyre'] = df['Tyre'].replace(['M', 'E', 'C', 'D', 'F' ,'G', 'P' ,'R'], 
                   ['Michelin', 'Englebert', 'Continental', 'Dunlop', 'Firestone',
                    'Goodyear', 'Pirelli', 'Rapson'])
```

```python
df.tail()
```

```python
df['Driver_nationality'] = df['Driver_nationality'].apply(lambda x : x[0])
```

```python
df['Team_nationality'] = df['Team_nationality'].apply(lambda x : x[0])
```

```python
df.tail()
```

```python
df.set_index('Year', inplace=True)
```

```python
df.to_csv('clean_data.csv')
```
